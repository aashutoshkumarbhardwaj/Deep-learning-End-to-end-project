```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder

import pickle
```

```python
data=pd.read_csv('Churn_Modelling.csv')
```

```python
data.head()
```

```python
#preprocessing

data=data.drop(['RowNumber','CustomerId','Surname'],axis=1)
```

```python
data
```

```python
#encode one categorical data

label_encoder=LabelEncoder()
data['Gender']=label_encoder.fit_transform(data['Gender'])
```

```python
data
```

```python
#onehot ecoding for geo

from sklearn.preprocessing import OneHotEncoder
onehotgeo=OneHotEncoder()
data_geo=onehotgeo.fit_transform(data[['Geography']])
data_geo
```

```python
onehotgeo.get_feature_names_out(['Geography'])
```

```python
geo_encodeed_pd=pd.DataFrame(data_geo.toarray(), columns=onehotgeo.get_feature_names_out(['Geography']))
geo_encodeed_pd
```

```python
##combines columns 

data=pd.concat([data.drop('Geography',axis=1),geo_encodeed_pd],axis=1)
```

```python
data.head()
```

```python
#save the encode adn scaler




with open('label_encoder.pkl','wb') as f:
    pickle.dump(label_encoder,f)

with open('onehot_encoder_geography.pkl','wb') as f:
    pickle.dump(onehotgeo,f)


```

```python
data.head()
```

```python
#divide data into dep. and indep 

x=data.drop('Exited',axis=1)
y=data['Exited']

#split data into train and test


x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)


##scalling these features

scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
x_test
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
import datetime
```

```python
model=Sequential([
    Dense(64,activation='relu',input_shape=(x_train.shape[1],)), #only written fro first layer as it is input layer
    Dropout(0.5),
    Dense(32,activation='relu'),
    Dropout(0.5),
    Dense(16,activation='relu'),
    Dropout(0.5),
    Dense(1,activation='sigmoid')
])
```

```python
model.summary()
```

```python
##compile the model
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
opt=tf.keras.optimizers.Adam(learning_rate=0.001)
loss=tf.keras.losses.BinaryCrossentropy()
model.compile(optimizer=opt,loss=loss,metrics=['accuracy'])
```

```python
#set up tensorboard

log_dir="logs/fit/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorboard_callback=TensorBoard(log_dir=log_dir, histogram_freq=1)
```

```python
tensorboard_callback=TensorBoard(log_dir=log_dir, histogram_freq=1)
```

```python
##early stopping

early_stopping=EarlyStopping(monitor='val_loss',patience=5,restore_best_weights=True)
```

```python
##training the model

history=model.fit(
    x_train,y_train,
    validation_data=(x_test,y_test),
    epochs=100,
    batch_size=32,
    callbacks=[early_stopping,tensorboard_callback] 
)
```

```python
model.save("model.h5")
```
