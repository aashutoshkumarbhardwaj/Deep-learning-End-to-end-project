```python
import tensorflow as tf
from tensorflow.keras.models import load_model
import pickle
import pandas as pd
import numpy as np
```

```python
#load trained model

model=load_model("model.h5")

with open("scaler.pkl","rb") as f:
    scaler=pickle.load(f)
++++++
with open("label_encoder.pkl","rb") as f:
    label_encoder=pickle.load(f)

with open("onehot_encoder_geography.pkl","rb") as f:
    one_hot_encoder_geography=pickle.load(f)

```

```python
#input data
input_data={
    "CreditScore":600,
    "Geography":"France",
    "Gender":"Male",
    "Age":40,
    "Tenure":3,
    "Balance":60000,
    "NumOfProducts":2,
    "HasCrCard":1,
    "IsActiveMember":1,
    "EstimatedSalary":50000
}
```

```python
geo_encoded=one_hot_encoder_geography.transform([[input_data["Geography"]]]).toarray()
geo_encoded_df=pd.DataFrame(geo_encoded,columns=one_hot_encoder_geography.get_feature_names_out(["Geography"]))
```

```python
geo_encoded_df
```

```python
import numpy as np
print(np.array(input_data).shape)
```

```python
input_data=pd.DataFrame(input_data)
input_data 
```

```python
input_df=pd.concat([input_data.reset_index(drop=True),geo_encoded_df],axis=1)
input_df
```

```python
input_data['Gender'] = label_encoder.transform([input_data['Gender']])[0]
input_df
```

```python
#concatenate input data with encoded features
input_df=pd.concat([input_data.reset_index(drop=True),geo_encoded_df],axis=1)
input_df
```

```python
print(input_df.shape)
print(input_df.columns)
print(len(input_df.columns))
```

```python
input_df = input_df.loc[:, ~input_df.columns.duplicated()]
```

```python
input_scaled = scaler.transform(input_df)
```

```python
#scalling the input data

input_scaled=scaler.transform(input_df)
input_scaled
```

```python
prediction=model.predict(input_scaled)
prediction
```
